```python
# Importing necerssary files
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , OneHotEncoder , StandardScaler 
```

# Task 1: Data splitting and Anti leakage setup

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
RANDOM_STATE = 42
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Select only columns with numbers
num_cols = df.select_dtypes(include=['number'])

# Calculate the correlation matrix for just these columns
corr_matrix = num_cols.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Numerical Features Only)')
plt.show()
```

```python
df =df.drop(columns=['CustomerId', 'RowNumber', 'Surname'])
```

```python
X = df.drop(columns=['Exited'])
y = df['Exited']
```

```python
y.value_counts()
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y ,test_size=42, random_state = RANDOM_STATE)
```

# Task 2: Feature Engineering & Preprocessing

```python
#Label Encoding of Gender
le = LabelEncoder()
X_train['Gender'] = le.fit_transform(X_train['Gender'])
X_test['Gender'] = le.transform(X_test['Gender'])
```

```python
# One-Hot Encoding of Geography
X_train = pd.get_dummies(X_train, columns=['Geography'], drop_first=True)
X_train.head()
```

```python
X_test = pd.get_dummies(X_test , columns=['Geography'], drop_first=True)
```

```python
# Finding Correlation
train_data = X_train.copy()
train_data['Exited']= y_train

# Finding Correlation
corr_matrix = train_data.corr()
print(corr_matrix)
```

```python
X_train['Balance_per_product'] = X_train['Balance']/X_train['NumOfProducts']
```

```python
X_train['Loyalty'] = X_train['Tenure']/X_train['Age']
```

```python
threshold = X_train['Balance'].quantile(0.75)
X_train['High_Value_Active'] = np.where((X_train['IsActiveMember']==1 &  (X_train['Balance'] >= threshold)), 1 , 0)
```

```python
# Scaling
scaler = StandardScaler()
num_cols = ['Balance_per_product', 'Loyalty' , 'High_Value_Active', 'CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
X_train[num_cols] = scaler.fit_transform(X_train[num_cols]) 
```

```python
X_test['Balance_per_product'] = X_test['Balance']/X_test['NumOfProducts']
X_test['Loyalty'] = X_test['Tenure']/X_test['Age']
X_test['High_Value_Active'] = np.where((X_test['IsActiveMember']==1 &  (X_test['Balance'] >= threshold)), 1 , 0)
X_test[num_cols] = scaler.fit_transform(X_test[num_cols]) 
```

# Task 3 : Model Development

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy = DummyClassifier(strategy='most_frequent')
# Training Dummy Classifier on Training dataset
dummy.fit(X_train , y_train)
# Dummy predicting on test dataset
dummy_pred = dummy.predict(X_test)
# Dummy Classifier Accuracy
dummy_acc = accuracy_score(y_test , dummy_pred)
dummy_acc
```

```python
# Decision Tree model 
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(max_depth = 3 , random_state= 42)
dt_model.fit(X_train , y_train)

dt_pred = dt_model.predict(X_test)
```

```python
dt_acc = accuracy_score(y_test , dt_pred)
dt_acc
```

```python
# Visualising the decision Tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plot_tree(dt_model, feature_names=X_train.columns, class_names=['Stay', 'Exit'], filled=True)
plt.show()
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100 , random_state=42)
rf.fit(X_train , y_train)
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test , rf_pred)
```

```python
rf_acc
```

```python
report = classification_report(y_test, rf_pred)

print("Classification Report:")
print(report)
```

```python
# XGBoost
from xgboost import XGBClassifier

# Initialize the model
xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, random_state=42)

# Train the model
xgb.fit(X_train, y_train)

# Make predictions
xgb_preds = xgb.predict(X_test)
```

# Task 4 : Hyperparameter tuning and cross-validation

```python
# Hypertune Random forest
from sklearn.model_selection import GridSearchCV

# Define the grid of parameters to search
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Best parameter values
print(grid_search.best_params_)
```

```python
# k-fold Cross Validation
from sklearn.model_selection import cross_val_score

# 1. Use the best model found by your GridSearch
best_rf_model = grid_search.best_estimator_

# 2. Perform 5-Fold Cross-Validation
cv_scores = cross_val_score(best_rf_model, X_train, y_train, cv=5)

# 3. View the results
print(f"All CV Scores: {cv_scores}")
print(f"Mean CV Accuracy: {cv_scores.mean():.4f}")
```

# Task 5 : Final Evaluation of Best Model

```python
# Use the best model to make final predictions
best_rf_pred = best_rf_model.predict(X_test)

# Calculate final accuracy
final_accuracy = accuracy_score(y_test, best_rf_pred)
print(f"Final Tuned Accuracy: {final_accuracy:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, best_rf_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Stay', 'Exit'])
disp.plot()
```

```python
# Generate the classification report
report = classification_report(y_test, best_rf_pred)

print("Classification Report:")
print(report)
```
